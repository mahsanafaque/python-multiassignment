```python
##Q3 : A movie theater charges different ticket prices depending on a person’s age. If a person is under the age
##of 3, the ticket is free; if they are between 3 and 12, the ticket is $10; and if they are over age 12, the
##ticket is $15. Write a loop in which you ask users their age, and then tell them the cost of their movie
##ticket.

##First input number of persons
##Then input age for all persons 
##Then check if age is below 3 and number of persons is only one then its will be print on screen only alone kid are not allow.
##if the kid is below 3 and the persons is more than one its will show you the total cost except 


print("Welcome to online ticket booking system\n\n\nWe have different prices of tickets according your age")

persons_number =int(input("\nHow many persons = "))


print("\nWe charge ticket cost according the age\n")
total_price = 0
for i in range(1,persons_number+1):
    age = int(input("\nPlease Enter Age "))
    if age <= 3 and persons_number != 1:
        print(" \nWe don't charge for below 3 years old.")
    elif age <= 3 and persons_number == 1:
        print(" \nWe don't allow alone kid on cinema.")    
    elif age <= 12 and age >= 3:
        print("We charged more than 3 years old and below 12 years old the ticket cost is 10$")
        total_price +=10
                
    elif age >= 12:
        print("We charged more than 12 years old the ticket cost is 12$")
        total_price +=12
        
print("\n\nYour Total Cost of Tickets is " + str(total_price)+"$")
```

```python
##Question4:
##Write a function called favorite_book() that accepts one parameter, title. The function should print a
##message, such as One of my favorite books is Alice in Wonderland. Call the function, making sure to
##include a book title as an argument in the function call.

print("Welcome to Read Online Books")
book_details = {
    "Ulysses": 
    {
        "Author":"James Joyce",
        "Publisher": "Sylvia Beach",
        "Publication date": "2 February 1922.",
        "Total Pages ":"730"
    },
    "Don Quixote": 
    {
        "Author":"Miguel de Cervantes",
        "Publisher": "Francisco de Robles",
        "Publication date": "1605 (Part One) 1615 (Part Two)",
        "Total Pages ":"863"
    },
    "Lolita": 
    {
        "Author":"Vladimir Nabokov",
        "Publisher": "Olympia Press",
        "Publication date": "1955",
        "Total Pages ":"336"
        
    }
}
input_book_name = input("Enter Your Book Name = ")

def find_book(input_book_name):
    flag = False 
    for book_key,book_values in book_details.items():
        if book_key == input_book_name:
            flag =  True
            break
    if(flag):            
        print("\nThanks for your input here is the details of your provided book name")
        print(str(book_key) + " " + str(book_values) )
    else:
        print("Thanks for your input but "+ input_book_name +" not available on store")    
    
            
find_book(input_book_name)
```

```python
##Question5:
##Guess the number game
##Write a program which randomly generate a number between 1 to 30 and ask the user in input field to guess the correct number. Give three chances to user guess the number and also give hint to user if hidden number is greater or smaller than the number he given to input field.

import random
print("WELCOME HERE TO CHECK YOUR LUCK")
total_score = 0
Flag = False
for x in range(0,3):
    number = random.randint(1, 30)
    user_number =input("Enter One Number which is between 1 to 30 ")
    if  str(user_number) == str(number):
        total_score += 10
        Flag = True
    if(Flag):
        print("Random Number is generated " + str(number) + " and Your number is "+ str(user_number))
        
    else:
        print("Your input is wrong and the random number was " + str(number))
print("Your total score is " + str(total_score))
```
